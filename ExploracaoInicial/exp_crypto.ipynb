```python
import yfinance as yf
from datetime import date
```

```python
import investpy as inv
```

```python
dic_list = inv.get_cryptos_dict()
dic_list
```

```python
dic = {f"{crypto['symbol']}-{crypto['currency']}" : {chave: valor for chave, valor in crypto.items() if chave != 'symbol'} for crypto in dic_list}
dic
```

```python
tickers = [i for i in dic]
tickers = tickers[0:50]
tickers
```

```python
TickersYF = yf.Tickers(tickers=tickers)
```

```python
df = TickersYF.download(period='1mo', interval='1wk', start='2018-01-01', end=date.today())
```

```python
df_history = df.stack(level=1).rename_axis(['Date', 'Ativo']).reset_index(level=1)
df_history
```
