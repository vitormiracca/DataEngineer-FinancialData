```python
tickers = {'ABCB4': 'ABCB4.SA', 'AGRO3': 'AGRO3.SA', 'RAIL3': 'RAIL3.SA', 'ALPA3': 'ALPA3.SA', 'ALPA4': 'ALPA4.SA', 'ALSO3': 'ALSO3.SA', 'AMAR3': 'AMAR3.SA', 'ABEV3': 'ABEV3.SA'}
```

```python
### HISTORICO PARA VARIOS

result_df = pd.DataFrame()

for t in tickers:
    TickerYF = yf.Ticker(tickers[t])
    df = TickerYF.history(period='1d', interval='1wk', start='2020-01-01', end=date.today())
    df['Ativo'] = t
    df = df[['Ativo', 'Open', 'Close', 'Low', 'High', 'Volume']]
    result_df = pd.concat([result_df, df])

result_df
```
