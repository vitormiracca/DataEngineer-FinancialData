### COtações

```python
tickers = {'ABCB4': 'ABCB4.SA', 'AGRO3': 'AGRO3.SA', 'RAIL3': 'RAIL3.SA', 'ALPA3': 'ALPA3.SA', 'ALPA4': 'ALPA4.SA', 'ALSO3': 'ALSO3.SA', 'AMAR3': 'AMAR3.SA', 'ABEV3': 'ABEV3.SA'}
```

```python
### HISTORICO PARA VARIOS

result_df = pd.DataFrame()

for t in tickers:
    TickerYF = yf.Ticker(tickers[t])
    df = TickerYF.history(period='1d', interval='1wk', start='2020-01-01', end=date.today())
    df['Ativo'] = t
    df = df[['Ativo', 'Open', 'Close', 'Low', 'High', 'Volume']]
    result_df = pd.concat([result_df, df])

result_df
```

```python
TickersYF = yf.Tickers(list(tickers.values()))
```

```python
TickersYF.tickers
```

```python
df_result = pd.DataFrame()

for i in TickersYF.tickers:
    ativo = i.replace(".SA", "")
    df_history = TickersYF.tickers[i].history(period='1d', interval='1wk', start='2020-01-01', end=date.today())
    df_history['Ativo'] = ativo
    df_history = df_history[['Ativo', 'Open', 'Close', 'Low', 'High', 'Volume']]
    df_result = pd.concat([df_result, df_history])

df_result
```

```python
########## MAIS PERFORMATICA #############

df_history = TickersYF.download(interval='1wk', start='2020-01-01', end=date.today(), period='1d')
df_history = df_history.stack(level=1).rename_axis(['Date', 'Ativo']).reset_index(level=1)
df_history = df_history[['Ativo', 'Open', 'Close', 'High', 'Low', 'Volume']]
df_history['Ativo'] = df_history.Ativo.replace(".SA", "", regex=True)
df_history
```
